    </div>
    <p style="color: #555; line-height: 1.6; margin: 0 0 15px 0; font-size: 13px;">
        Based on <a href="https://en.wikipedia.org/wiki/Three-act_structure" style="color: #17a2b8;">Aristotle's Poetics</a>
        and refined by <a href="https://en.wikipedia.org/wiki/Syd_Field" style="color: #17a2b8;">Syd Field</a>,
        the three-act structure creates compelling narratives:
    </p>
    <div style="display: grid; grid-template-columns: repeat(3, 1fr); gap: 15px;">
        <div style="background: #e8f5e9; padding: 12px; border-radius: 5px; border-top: 3px solid #28a745;">
            <h4 style="color: #28a745; font-size: 12px; margin: 0 0 8px 0;">Act 1: Setup (Context)</h4>
            <ul style="color: #555; font-size: 11px; margin: 0; padding-left: 15px;">
                <li>Establish baseline</li>
                <li>Introduce the problem</li>
                <li>Set expectations</li>
            </ul>
        </div>
        <div style="background: #fff3e0; padding: 12px; border-radius: 5px; border-top: 3px solid #ff9800;">
            <h4 style="color: #ff9800; font-size: 12px; margin: 0 0 8px 0;">Act 2: Conflict (Complication)</h4>
            <ul style="color: #555; font-size: 11px; margin: 0; padding-left: 15px;">
                <li>Show the challenge</li>
                <li>Reveal patterns</li>
                <li>Build tension</li>
            </ul>
        </div>
        <div style="background: #f3e5f5; padding: 12px; border-radius: 5px; border-top: 3px solid #9c27b0;">
            <h4 style="color: #9c27b0; font-size: 12px; margin: 0 0 8px 0;">Act 3: Resolution (Conclusion)</h4>
            <ul style="color: #555; font-size: 11px; margin: 0; padding-left: 15px;">
                <li>Present solution</li>
                <li>Show impact</li>
                <li>Call to action</li>
            </ul>
        </div>
    </div>
</div>


```python
# ============================================
# CELL 3: Demonstrating Three-Act Structure
# Purpose: Show how to structure a data story
# ============================================

# Generate data for our story
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Act 1: Stable baseline
customer_satisfaction = [85, 84, 85, 83, 84, 82, 78, 72, 68, 65, 70, 78]

# Create the three-act visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Act 1: Setup - Show the baseline
axes[0].plot(months[:6], customer_satisfaction[:6], 'o-', color='#28a745', linewidth=2, markersize=8)
axes[0].set_title('Act 1: The Setup\n"Everything Seemed Fine"', fontweight='bold', fontsize=12)
axes[0].set_ylim([50, 100])
axes[0].axhline(y=80, color='gray', linestyle='--', alpha=0.5, label='Target: 80%')
axes[0].set_ylabel('Customer Satisfaction (%)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Act 2: Conflict - Show the problem
axes[1].plot(months[5:10], customer_satisfaction[5:10], 'o-', color='#dc3545', linewidth=2, markersize=8)
axes[1].set_title('Act 2: The Conflict\n"Crisis Emerges"', fontweight='bold', fontsize=12)
axes[1].set_ylim([50, 100])
axes[1].axhline(y=80, color='gray', linestyle='--', alpha=0.5)
axes[1].axhline(y=70, color='red', linestyle='--', alpha=0.5, label='Critical: 70%')
axes[1].annotate('Problem!', xy=(months[8], 68), xytext=(months[7], 60),
                arrowprops=dict(arrowstyle='->', color='red', lw=2))
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Act 3: Resolution - Show the recovery
axes[2].plot(months[9:], customer_satisfaction[9:], 'o-', color='#17a2b8', linewidth=2, markersize=8)
axes[2].set_title('Act 3: The Resolution\n"Recovery Through Action"', fontweight='bold', fontsize=12)
axes[2].set_ylim([50, 100])
axes[2].axhline(y=80, color='gray', linestyle='--', alpha=0.5, label='Target: 80%')
axes[2].annotate('Recovery!', xy=(months[11], 78), xytext=(months[10], 85),
                arrowprops=dict(arrowstyle='->', color='green', lw=2))
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.suptitle('Customer Satisfaction Story: Crisis and Recovery', fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()

pretty_print("Three-Act Structure Applied",
             "Notice how breaking the story into three acts creates drama and maintains attention. Each act has a clear purpose.",
             'info')
```

```python
# ============================================
# HANDS-ON EXERCISE 1: Build Three-Act Story
# Task: Create narrative from sales data
# ============================================

# Sales data telling a story
quarters = ['Q1-19', 'Q2-19', 'Q3-19', 'Q4-19', 'Q1-20', 'Q2-20', 'Q3-20', 'Q4-20']
sales = [120, 125, 122, 118, 95, 88, 102, 135]  # Notice the pattern!

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Act 1 - Stable business (Q1-19 to Q4-19)
axes[0].plot(quarters[:4], sales[:4], 'o-', color='#28a745', linewidth=2.5, markersize=8)
axes[0].set_title('Act 1: Steady Growth\n"Business as Usual"', fontweight='bold')
axes[0].set_ylabel('Sales ($M)')
axes[0].set_ylim([70, 150])
axes[0].grid(True, alpha=0.3)

# Act 2 - Crisis hits (Q4-19 to Q2-20)
axes[1].plot(quarters[3:6], sales[3:6], 'o-', color='#dc3545', linewidth=2.5, markersize=8)
axes[1].set_title('Act 2: The Crisis\n"Pandemic Impact"', fontweight='bold')
axes[1].set_ylim([70, 150])
axes[1].annotate('Lockdown!', xy=(quarters[5], sales[5]), xytext=(quarters[4], 75),
                arrowprops=dict(arrowstyle='->', color='red', lw=2))
axes[1].grid(True, alpha=0.3)

# Act 3 - Recovery and growth (Q2-20 to Q4-20)
axes[2].plot(quarters[5:], sales[5:], 'o-', color='#17a2b8', linewidth=2.5, markersize=8)
axes[2].set_title('Act 3: The Comeback\n"Digital Transformation Success"', fontweight='bold')
axes[2].set_ylim([70, 150])
axes[2].annotate('Record High!', xy=(quarters[7], sales[7]), xytext=(quarters[6], 145),
                arrowprops=dict(arrowstyle='->', color='green', lw=2))
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"📊 {student_name}, uncomment the code to reveal the three-act sales story!")

pretty_print("Story Complete!",
             "Excellent! You've transformed raw sales data into a compelling narrative of crisis and recovery.",
             'success')
```

<div style="background: white; padding: 15px 20px; margin-top: 20px; border-left: 3px solid #17a2b8;">
    <h3 style="color: #17a2b8; font-size: 14px; margin: 0 0 8px 0;">The Inverted Pyramid Structure</h3>
    <p style="color: #555; line-height: 1.6; margin: 0; font-size: 13px;">
        Borrowed from <a href="https://en.wikipedia.org/wiki/Inverted_pyramid_(journalism)" style="color: #17a2b8;">journalism</a>,
        the inverted pyramid presents the most important information first. In data visualization, this means:
        <strong>Lead with the insight</strong>, then provide supporting details, and finally show the methodology.
        This approach respects your audience's time and ensures key messages aren't missed.
    </p>
</div>


<div style="background:#f8fffe; border:1px solid #17a2b8; border-radius:10px; padding:18px; margin:18px 0;">
  <div style="background:linear-gradient(90deg,#17a2b8 0%, #14919e 100%); color:white; padding:10px 14px; margin:-18px -18px 14px -18px; border-radius:10px 10px 0 0;">
    <h3 style="margin:0; font-size:14px; letter-spacing:0.8px; text-transform:uppercase;">What is the Progressive Disclosure Method?</h3>
  </div>

  <p style="color:#333; font-size:13px; line-height:1.6; margin:0 0 10px 0;">
    <strong>Progressive disclosure</strong> is a storytelling technique where you reveal information in deliberate stages—starting with the most digestible, high-level view and then adding detail only when it becomes useful. The goal is to reduce cognitive load, prevent chart clutter, and guide attention toward the insight you want the audience to reach.
  </p>

  <div style="display:grid; grid-template-columns: repeat(3, 1fr); gap:12px; margin-top:10px;">
    <div style="background:#ffffff; border-left:3px solid #17a2b8; border-radius:6px; padding:10px;">
      <h4 style="margin:0 0 6px 0; font-size:12px; color:#17a2b8;">Why it Works</h4>
      <ul style="margin:0; padding-left:16px; color:#555; font-size:12px; line-height:1.5;">
        <li>Reduces cognitive overload</li>
        <li>Creates a natural narrative arc</li>
        <li>Keeps focus on the “so what”</li>
      </ul>
    </div>
    <div style="background:#ffffff; border-left:3px solid #28a745; border-radius:6px; padding:10px;">
      <h4 style="margin:0 0 6px 0; font-size:12px; color:#28a745;">Core Pattern</h4>
      <ol style="margin:0; padding-left:16px; color:#555; font-size:12px; line-height:1.5;">
        <li><strong>Stage 1:</strong> Show the headline metric or trend.</li>
        <li><strong>Stage 2:</strong> Add essential drivers or constraints.</li>
        <li><strong>Stage 3:</strong> Reveal the underlying mechanism or takeaway.</li>
      </ol>
    </div>
    <div style="background:#ffffff; border-left:3px solid #ffc107; border-radius:6px; padding:10px;">
      <h4 style="margin:0 0 6px 0; font-size:12px; color:#e0a800;">Do’s & Don’ts</h4>
      <ul style="margin:0; padding-left:16px; color:#555; font-size:12px; line-height:1.5;">
        <li><strong>Do</strong> start simple; add only what’s needed.</li>
        <li><strong>Do</strong> annotate transitions (“Now add costs…”).</li>
        <li><strong>Don’t</strong> reveal everything at once.</li>
        <li><strong>Don’t</strong> mix too many encodings per stage.</li>
      </ul>
    </div>
  </div>

  <div style="margin-top:12px; background:#f0fbfd; border:1px dashed #17a2b8; border-radius:6px; padding:10px;">
    <p style="margin:0; color:#0e5a63; font-size:12px;">
      In the next cell, we implement this pattern in three stages: <em>Revenue only → Revenue + Costs → Profit emphasis</em>.
    </p>
  </div>
</div>


```python
# ============================================
# CELL 4: Progressive Disclosure Technique
# Purpose: Show how to reveal information gradually
# ============================================

# Data for progressive disclosure
years = list(range(2015, 2024))
revenue = [45, 52, 61, 68, 72, 78, 85, 92, 98]
costs = [40, 44, 48, 51, 53, 55, 58, 60, 62]
profit = [r - c for r, c in zip(revenue, costs)]

# Create progressive disclosure in 3 stages
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Stage 1: Just revenue
axes[0].plot(years, revenue, 'o-', color='#28a745', linewidth=2.5, label='Revenue')
axes[0].set_title('Stage 1: The Good News', fontweight='bold')
axes[0].set_ylabel('Amount ($M)')
axes[0].set_ylim([0, 100])
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Stage 2: Add costs
axes[1].plot(years, revenue, 'o-', color='#28a745', linewidth=2.5, label='Revenue')
axes[1].plot(years, costs, 's-', color='#dc3545', linewidth=2.5, label='Costs')
axes[1].set_title('Stage 2: The Full Picture', fontweight='bold')
axes[1].set_ylim([0, 100])
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Stage 3: Highlight profit
axes[2].bar(years, profit, color='#17a2b8', alpha=0.7, label='Profit')
axes[2].plot(years, profit, 'o-', color='#FF6B6B', linewidth=2.5, markersize=8)
axes[2].set_title('Stage 3: The Real Story', fontweight='bold')
axes[2].set_ylabel('Profit ($M)')
axes[2].legend()
axes[2].grid(True, alpha=0.3, axis='y')

plt.suptitle('Progressive Disclosure: Building Understanding Step by Step',
            fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()

pretty_print("Progressive Disclosure",
             "By revealing information gradually, we guide the viewer's understanding and maintain engagement.",
             'info')
```

```python
# ============================================
# HANDS-ON EXERCISE 2: Progressive Disclosure
# Task: Reveal market share story in stages
# ============================================

# Market share data
companies = ['Our Company', 'Competitor A', 'Competitor B', 'Others']
market_2022 = [15, 35, 30, 20]
market_2023 = [22, 32, 28, 18]
market_2024 = [31, 28, 25, 16]

# Create 4 subplots for progressive disclosure
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Stage 1 - Just show 2022 as baseline
axes[0].pie(market_2022, labels=companies, autopct='%1.1f%%', startangle=90)
axes[0].set_title('2022: The Starting Point', fontweight='bold')

# Stage 2 - Show 2023 growth
axes[1].pie(market_2023, labels=companies, autopct='%1.1f%%', startangle=90,
           colors=['#28a745', '#95a5a6', '#95a5a6', '#95a5a6'])
axes[1].set_title('2023: Gaining Ground', fontweight='bold')

# Stage 3 - Show 2024 leadership
axes[2].pie(market_2024, labels=companies, autopct='%1.1f%%', startangle=90,
           colors=['#FFD700', '#95a5a6', '#95a5a6', '#95a5a6'])
axes[2].set_title('2024: Market Leader!', fontweight='bold')

# Stage 4 - Show the trend
years = ['2022', '2023', '2024']
our_growth = [15, 22, 31]
axes[3].plot(years, our_growth, 'o-', color='#FFD700', linewidth=3, markersize=10)
axes[3].set_title('The Growth Story', fontweight='bold')
axes[3].set_ylabel('Market Share (%)')
axes[3].set_ylim([10, 35])
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"📈 {student_name}, reveal the market dominance story by uncommenting the code!")

pretty_print("Progressive Story Complete!",
             "Perfect! You've shown how to build understanding and excitement through gradual revelation.",
             'success')
```

```python
# ============================================
# CELL 5: Before/After Comparison Narratives
# Purpose: Show transformation stories
# ============================================

# Performance metrics before and after intervention
metrics = ['Efficiency', 'Quality', 'Speed', 'Cost Savings', 'Customer Satisfaction']
before_scores = [45, 52, 48, 35, 58]
after_scores = [78, 85, 82, 72, 88]

# Create compelling before/after visualization
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5), constrained_layout=True)

# Before state
max_val = max(before_scores) * 1.15  # add margin for labels
ax1.barh(metrics, before_scores, color='#dc3545', alpha=0.7)
ax1.set_xlim([0, max_val])
ax1.set_title('BEFORE: Struggling Performance', fontweight='bold')
ax1.set_xlabel('Score')
ax1.grid(False)
for i, v in enumerate(before_scores):
    ax1.text(v + max_val * 0.01, i, str(v), va='center', ha='left')

# After state
max_val = max(after_scores) * 1.15
ax2.barh(metrics, after_scores, color='#28a745', alpha=0.7)
ax2.set_xlim([0, max_val])
ax2.set_title('AFTER: Transformed Results', fontweight='bold')
ax2.set_xlabel('Score')
ax2.grid(False)
for i, v in enumerate(after_scores):
    ax2.text(v + max_val * 0.01, i, str(v), va='center', ha='left')

# The improvement story
improvements = [a - b for a, b in zip(after_scores, before_scores)]
max_val = max(improvements) * 1.2
colors = ['#FFD700' if i >= 30 else '#17a2b8' for i in improvements]
ax3.barh(metrics, improvements, color=colors)
ax3.set_xlim([0, max_val])
ax3.set_title('THE IMPACT: Dramatic Improvements', fontweight='bold')
ax3.set_xlabel('Improvement Points')
ax3.grid(False)
for i, v in enumerate(improvements):
    ax3.text(v + max_val * 0.01, i, f'+{v}', va='center', ha='left', fontweight='bold')

plt.suptitle('Digital Transformation Success Story', fontsize=16, fontweight='bold', y=1.05)
plt.show()

pretty_print("Before/After Narratives",
             "Transformation stories are powerful. They show possibility and inspire action.",
             'info')
```
